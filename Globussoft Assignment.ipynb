```python
## Question: - 4
import math

@staticmethod
def main():
    n = None
    m = None
    a = None
    b = None
    s = None
    while scanf("%lf%lf%lf%d%d", a, b, s, m, n)!=EOF:
        if a+b+s+m+n == 0:
            break

        angel = math.atan2(b *n,a *m)*180/math.acos(-1.0)
        v = math.sqrt(a *a *m *m+b *b *n *n)/s
        print("{0:.2f} {1:.2f}\n".format(angel, v), end = '')

class DefineConstants(object):
    INF = 0x3f3f3f3f
```
